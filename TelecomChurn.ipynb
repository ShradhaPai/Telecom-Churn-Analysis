Importing the necessary libraries to analyse the dataset like pandas for manipulation and analysis, matplotlib and seaborn for visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the dataset and displaying the first few rows of the datatset

```python
df = pd.read_csv('Telco-Customer-Churn.csv')
print(df.head())
```

Finding the number of rows and columns in the dataset

```python
df.shape
```

Displaying all the column names in the dataset

Counting the number of male and female subscribers

```python
df.groupby('gender').size()
```

Plotting Churn by Contarct

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'Contract', hue ='Churn', data = df)
plt.show()
```

Here we can see that subscribers who opted for month to month contract were most likely to churn rather than the yearly contract ones. As the tenure of contract increases the subscriber is less likely to churn.

Plotting Churn by Payment Method

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'PaymentMethod', hue ='Churn', data = df)
plt.show()
```

From this plot one can observe that the subscribers who chose electronic check as mode of payment had the highest churn rate and as the payment method got automatic customers were less likely to churn.

Plotting Churn by InternetService

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'InternetService', hue ='Churn', data = df)
plt.show()
```

This shows subscribers who opted for fiber optic internet service had the highest churn rate than DSL and no internet service.

Plotting Churn by Partner

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'Partner', hue ='Churn', data = df)
plt.show()
```

From the plot its notable that subscribers who did have a partner were more prone to churn behaviour.

Plotting Churn by Dependents

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'Dependents', hue ='Churn', data = df)
plt.show()
```

Its seen that when the subscribers had dependents it was highy unlikely for them to churn.

Plotting Churn by TechSupport

```python
plt.figure(figsize=(10,5))
sns.countplot(x = 'TechSupport', hue ='Churn', data = df)
plt.show()
```

When the subscribers had tech support they were less likely to churn and those who did not have it were more likely to churn.
